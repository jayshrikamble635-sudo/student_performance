```python
import streamlit as st

# Title
st.title("Student Performance Predictor")

# Inputs
hours = st.number_input("Hours Studied", min_value=0.0, max_value=24.0, step=0.5)
attendance = st.slider("Attendance Percentage", 0, 100, 50)
sleep = st.number_input("Sleep Hours", min_value=0.0, max_value=12.0, step=0.5)

# Prediction function
def predict_performance(hours, attendance, sleep):
    score = (hours * 10) + (attendance * 0.5) + (sleep * 5)
    return score

# Button
if st.button("Predict"):
    result = predict_performance(hours, attendance, sleep)
    st.write("Predicted Performance Score:", round(result, 2))
```
